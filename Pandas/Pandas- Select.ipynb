## Pandas -- Select Data

There are different methods to select data from a dataframe.

```python
import pandas as pd
fp = r'C:\Users\Gokul G\Desktop\WORK\GISISH\GIS-ish\data\data_missing.csv'
data = pd.read_csv(fp,sep=',',  na_values=[-999.0,-999,0])
```

When you have a large amount of data you need to seperate the necessary subset of data. There are different methods to select data from a dataframe. We can select specific rows, columns or individual values.

## Basic

The most basic method of selecting data is passing a the column or a list of columns through  "[]" to select columns in that order. If a column is not contained in the DataFrame, an exception will be raised. 

If we use the name of column to select data. For a single column to be selected we can pass that through the syntax.

```python
data['AOD_500nm']
```

But for multiple columns, the names of columns have to be passed as a list itself.

```python
data1 = data[["AOD_500nm",'Precipitable_Water(cm)', '380-500_Angstrom_Exponent','NUM_DAYS[AOD_500nm]','Latitude(degrees)', 'Longitude(degrees)', 'Elevation(meters)']]
data1
```

In this basic method we can also slice data. But it will only slice the rows. You can select the specific columns [0:11] and even decide to skip the odd ones in between.

```python
data2= data[0:11:2]
data2
```

## loc()

For selection the pandas loc() function can be used in two ways.

- To select using label/index.
- To select using a conditional statement.

The arguments for loc() are row selector and column selector. This function can be used to select one row or slice of row by calling the index of the row. To print the single row neatly add a bracket again.



```python
data.loc[7]
```

```python
data.loc[[5]]
```

```python
data.loc[0:5]
```

```python
data.loc[:,['Month','AOD_1020nm']]
```

To select specific columns we have to set the row selector first then set the columns selector. Single column can be called separately, but multiple column needs to be called as a list. The specific rows can be sliced according to the need as the row selector.



```python
data.loc[2:10,['Month','AOD_1020nm']]
```

Now the selection of columns using conditional statements is one of the most common use of loc() function. This is also called filtering of data.

In most use cases, you will make selections based on the values of different columns in your data set. if we use the condition of only one column.



```python
data.loc[data['AOD_500nm'] < 0.38]	
```

In the case of multiple conditions the “AND” (|) and “OR” (&) operators are used for selection.



```python
data.loc[(data['AOD_500nm'] < 0.42) & (data['NUM_DAYS[AOD_500nm]'] > 20)]
```

## iloc()

Similarly iloc() is used for purely integer-location based indexing for selection by position.


```python
data.iloc[0:5:, 0:2]
```

## at()

To access a single value in a dataframe at() function is used. this function is similar to loc() function, it can use labels and column names for selecting data. Similar to iloc(), iat() function selects data based on position indexes.

```python
data.at[5,'AOD_500nm']
```

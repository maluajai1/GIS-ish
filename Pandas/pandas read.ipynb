# Reading data

Data can be provided in many different formats (CSV, TSV, JSON, HTML, SAS, etc.) reading them is the first step to the data analysis.

To read a comma-separated values(csv) data file, and store the contents of that file into a variable called data we use the pandas.read_csv() function:

read_csv() is used to read a csv file into DataFrame, commonly used  variables are "df" and "data"

```python
fp = r'C:\Users\Gokul G\Desktop\WORK\GISISH\GIS-ish\data\Kanpur.csv'
data = pd.read_csv(fp,sep=',',index_col= "Month", skiprows=6, na_values=[-999.0,-999])
```

Some common arguments that we need to know for reading files are:

- fp - file path

If the file to be used is in the same location as the python file then only relative  location is needed to be given as the file path(fp), if the csv is in another location pass in the full location to the file path. 
In some cases the location can be called without the ''r(raw strings)'' in the beginning, but the r present will provide an escape sequence and it will properly read every file path, so it is always safer to put an ''r'' before the file path.

- sep 

This parameter is used to indicate that we are dealing with a comma-delimited text file. Comma is also the default value for the sep-parameter. Depending on your operating system and the input data, you might need to specify other separators such as ''tab'' and ";".

- skiprows

This is to indicate the line numbers to skip (0-indexed) or number of lines to skip (int) at the start of the file. 

- index_col

Column(s) to use as the row labels of the DataFrame, either given as string name or column index. "index_col = False" can be used to force pandas to not use the first column as the index. 

- usecols

This is used when only a subset of the columns is needed for the analysis. Columns are called either by positional (i.e. integer indices into the document columns) or strings that correspond to column names provided either by the user in names or inferred from the document header row(s).

- na_values

This is used to recognize the common null values or no value available. Identifying the null values before hand helps to make analysis better. If dict passed, specific per-column NA values.

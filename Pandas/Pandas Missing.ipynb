## Pandas -- Missing Data

Real world data will have values missing in different columns or rows, thus we have to clean our data before proceeding to any analysis. Cleaning our data is one of the important initial step.

In a dataframe missing data can occur when no information is provided for one or more items or for a whole unit. Missing Data can also refer to as NA(Not Available) values in pandas. Let look in to all the methods we use to clean our data from NaN values.

```python
import pandas as pd
fp = r'C:\Users\Gokul G\Desktop\WORK\GISISH\GIS-ish\data\Kanpur.csv'
data = pd.read_csv(fp,sep=',',index_col= "Month", skiprows=6, na_values=[-999.0,-999,0])
```


## isna(), notna(), isnull(), notnull()

First we have to identify the missing values in our data for this we have different functions like isna(), notna(), isnull(), notnull(). These functions help in checking whether a value is NaN or not.

These function takes a scalar or array-like object and indicates whether values are missing.

isna() and isnull() return a boolean mask where True is set for missing values (NaN) with False for non-missing values. I researched the difference between both these function and found that they are identical ( Shocking! I know). The best explanation was here.

```python
data.isna()
```


The reverse of this is done by isnull() and notnull() they return a boolean mask where False is set for missing values (NaN) with True for non-missing values.

```python
data.notnull()
```

```python
data.dropna(axis=1, thresh=200, inplace=True)
data
```

After finding the missing values we need to clean them. Pandas objects are equipped with various data manipulation methods for dealing with missing data.

## dropna()

This function is used to remove NA values. Lets look into all the arguments of the fuinction.

```python
data.dropna(axis=0,thresh=10, inplace=True)
data
```

To choose whether we need to delete the rows or columns containing the Na values we use axis. only one can be deleted at a time.

- If axis is set as “0” or ‘index’ the functions Drop rows which contain missing values.
- If axis is set as “1” or ‘column’ the functions Drop columns which contain missing values.

To determine the condition in which to delete the row or column we use how argument.

- When how is set to ‘any’, drop that row or column if any NA values are present.
- When how is set to ‘all’ , drop that row or column if all values are NA.

In the cases where we cannot choose any or all we use thresh argument. Here we set an integer value as we require that many non-NA values. Cannot be combined with how.

If we need to delete the rows for some specific column of missing values then we use subset argument. Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.

Now the last argument inplace is to decide if we should change the parent dataframe or create new one.

- If inplace is set to True then the dataframe will be permanently changes
- If inplace is set to False then the change will not be made on the parent dataframe
